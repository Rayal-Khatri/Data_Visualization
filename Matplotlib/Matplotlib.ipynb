# Dependencies

```python
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
import random
import pandas as pd
```

# Line Plot

```python

days = list(range(1, 16))

ktm_tempe = [22.5, 23.0, 21.5, 20.0, 19.5, 18.0, 17.5, 
               18.5, 19.0, 20.0, 21.0, 22.0, 23.0, 24.0, 25.0]

lal_temp = [21.5, 22.0, 22.8, 23.5, 24.0, 24.5, 25.0, 25.5, 26.0, 26.5, 27.0, 27.5, 28.0, 28.5, 29.0]
plt.axis([0,20,15,30])     #adding Axis

plt.plot(
        days, 
        ktm_tempe,
        # color = 'g', 
        # marker = 'o', 
        # linestyle = ':',
        'go:', 
        linewidth =1,
        markersize = 5,
        label = 'Kathmandu'
    )     #plotting

plt.plot(
    days,
    lal_temp,
    'ro--',
    linewidth = 1,
    markersize = 5,
    label = 'Lalitpur'
)

style.use("ggplot")


plt.title("Nepal Temperature", fontsize=20)      #Title
plt.xlabel("Day",fontsize= 13)               #Naming x axis
plt.ylabel('Temperature',fontsize= 13)        
plt.grid( linestyle=":",linewidth= 2)       #grid 

plt.legend(loc = 4)   #line defination and location of the line defination

```

# Histogram

```python
ml_students_age = np.random.randint(18,45,(100))
py_students_age = np.random.randint(13,45,(100))
```

```python
bins = [15,20,25,30,35,40,45]

style.use("ggplot")

plt.figure(figsize=(16,9))
plt.hist(
        [ml_students_age,py_students_age],
        bins,
        rwidth= .8,
        histtype="bar",
        # orientation="horizontal",
        color=['r','g'],
        label=["Ml_Students",'PY_Students']
    ) 

# plt.hist(
#         py_students_age,
#         bins,
#         rwidth= .8,
#         histtype="bar",
#         # orientation="horizontal",
#         color='g',
#         label="PY_Students"
#     ) 
plt.title("Students Age",fontsize= 20)
plt.xlabel("Age")
plt.ylabel("No. of students")
plt.legend()
```

# Bar Chart

```python
classes = ["Python", "R", "AI", "ML", "DS"]
class1_students = [30, 10, 20, 25, 10] # out of 100 student in each class
class2_students = [40, 5, 20, 20, 10]
class3_students = [35, 5, 30, 15, 15] 
width = 0.2
classes_index = np.arange(len(classes))
```

```python
style.use("bmh")
plt.figure(figsize=(16,9))

plt.barh(
    classes_index, 
    class1_students, 
    width,
    color = 'y',
    edgecolor = 'b',
    linewidth = 2,
    alpha = .9,
    label = 'Class_1',
)
plt.barh(
    classes_index +width + 0.05, 
    class2_students, 
    width,
    align="center",
    color = 'r',
    edgecolor = 'k',
    linewidth = 2,
    alpha = .9,
    label = 'Class_2',
)
plt.barh(
    classes_index + width + width+0.1, 
    class3_students, 
    width,
    align="center",
    color = 'g',
    edgecolor = 'b',
    linewidth = 2,
    alpha = .9,
    label = 'Class_3',
)


plt.yticks(
    classes_index +width,
    classes,
    rotation = 20,
    )

# plt.barh(classes, class1_students)  for horizontal bar
plt.legend()
plt.margins(.01)        #Margin
plt.title("Classes & Subjects",fontsize = 40)
plt.xlabel('Classes',fontsize = 20)
plt.ylabel('Students',fontsize = 20)

```

# Pie Chart

```python
classes = ["Python", 'R', 'Machine Learning', 'Artificial Intelligence', 'Data Sciece']
class1_students = [45, 15, 35, 25, 30] 
plt.figure(figsize=(16,9))
textprops = {"fontsize" : 15}
wedgeprops = {"linewidth": 4, "width": 1, "edgecolor":'k'}
plt.title("Class")
plt.pie(
            class1_students,
            labels=classes,
            explode=[.05,0,0.1,0,0],
            colors=['r','b','y','g','m'],
            autopct=  '%0.1f%%',            # to show the percentage
            pctdistance= .6,
            # labeldistance= 1.4,       # to set up a label distance
            # counterclock= True,
            shadow= True,
            radius=1.1,
            startangle=270,             # decide at what angle the pie chart starts
            textprops= textprops,
            wedgeprops= wedgeprops,
            # center=(2,3),               #decide the position of pie chart
            # frame=True,
            # rotatelabels=True
        )
plt.legend(loc= 2)
plt.show()
```
