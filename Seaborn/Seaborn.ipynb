# Histogram 

```python
plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
sns.histplot(tips_df['size'])

plt.subplot(2, 2, 2)
sns.histplot(tips_df['tip'])

plt.subplot(2, 2, 3)
sns.histplot(tips_df['total_bill'])

plt.subplot(2, 2, 4)
plt.pie([1])

plt.tight_layout()
plt.show()
```

```python
sns.displot(
        tips_df["total_bill"],
        bins = 10,
        kde = True,
        rug= True,
        color='r',
        label = "Total Bill",
    )

plt.legend()
```

```python
plt.figure(figsize=(16,9))
sns.set()

tips_df.total_bill.sort_values()

bins = [0,5,10,15,20,25,30,35,40,45,50,55]

sns.histplot(
                tips_df['total_bill'],
                bins = bins,
                kde = True,
                line_kws= {
                    'color' : '#DC143C'
                }
             )


plt.xticks(bins)

plt.title("Histogram of Total BIll")
plt.legend() 
```

# Barplot/ Bar Graph

```python
order = ['Sun', 'Thur', 'Fri','Sat']
kwargs = {
        'alpha': 0.9, 
        'linestyle' : '--',
        'linewidth' : 2,
        "edgecolor" : 'k',
    }


sns.barplot(
    # x = tips_df.day,     
    x= 'day',
    y = 'total_bill',
    hue= 'sex',
    data = tips_df,
    order= order,
    hue_order=['Female',"Male"],
    estimator= np.max,
    # errorbar=('ci', 100),
    n_boot= 2,
    saturation= 10,
    **kwargs
)
```

```python
plt.figure(figsize=(16,9))
sns.barplot(
    x= 'day',
    y = 'total_bill',
    hue= 'sex',
    data = tips_df,
    order= order,
    hue_order=['Female',"Male"],
    # n_boot= 2,
    saturation= 10,
    **kwargs
    )


plt.title ("Tips DataFrame")
plt.xlabel('Days')
plt.ylabel('Total Bill'),    
plt.show()
```
